## Series
Series are similar to lists in python. They are single dimensional data structures and generally indexed starting on Zero.
They are generated by passing python list in the constructor

```python
airports = pd.Series(['Kathmandu','Pokhara','Biratnagar','Bharatpur','Joomla'])
print(airports)
```

```python
airports[2]
for val in airports:
    print(val)
```

## DataFrame
A dataframe is a 2 dimensional data structure, with columns and rows.
It takes list of lists as input. The column names could be added at the end of the constructor

```python
airports = pd.DataFrame([
    ['Kathmandu','Bagmati','NP'],
    ['Pokhara','','NP'],
    ['Bharatpur','','NP']
])
print(airports)
```

```python
airports
```

Adding column names

```python
airports = pd.DataFrame(
[
    ['Kathmandu','Bagmati','NP'],
    ['Pokhara','','NP'],
    ['Bharatpur','','NP']
],
    columns=['Airport name','City','Country']
)
print(airports)
```

```python
airports[2:]
```

```python
airports
```

```python
for index,row in airports.iterrows():
    print(index, row['Airport name'], row['Country'])
```

```python
airports['Airport name']
airports[['Airport name']]
airports[['Airport name','Country']]

```

```python
airports.iloc[0,0]
airports.iloc[2,0]
```

```python
airports.iloc[:,:] #all values from rows or columns
```

```python
airports.iloc[0:2,:] #get first two rows only. 
```

```python
airports.iloc[:,0:2] #get first two cols only. last index is not included
```

```python
airports.iloc[:,[0,2]] #only two cols
airports.loc[:,['Airport name','Country']] #get two cols by name
```
